### ۳.۴ الگوریتم *Weighted A

الگوریتم ***Weighted A** نسخه‌ای از الگوریتم *A است که در آن از یک **وزن (w)** برای تسریع فرایند جستجو استفاده می‌کنیم.  
هدف از اضافه کردن این وزن، کاهش زمان جستجو  می‌باشد.

تابع هزینه در الگوریتم *Weighted A به صورت زیر تعریف می‌شود:

$
f(n) = g(n) + h(n) \times w
$

که در آن:

-  $g(n)$ : هزینه واقعی از مبدأ تا گره $n$ 
-  $h(n)$ : مقدار تخمینی فاصله از  $n$  تا مقصد (تابع $heuristic$)
-  $w$ : ضریب وزن (عدد مثبت، معمولاً بین 1 تا 5)

> اگر مقدار  $1 = w$  باشد، الگوریتم دقیقاً همان *A استاندارد است.  
> هر چه  $w$  بزرگ‌تر شود، تأثیر heuristic در تصمیم‌گیری بیشتر و جستجو سریع‌تر .
